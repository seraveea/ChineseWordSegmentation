```python
def get_new_word(news_dic,p_dic):
    for i in news_dic:
        if i not in p_dic:
            new_word.append(i)
    return new_word

def filter_length(dic,length):
    res = []
    for i in dic:
        if len(i)== length:
            res.append(i)
    return res
```
